```python
# drop na values
df.dropna(inplace = True)
# Split the first column "Audience List" by the deliminator " » "
dfa = df["Audience List"].str.split(pat=" » ", n = 4, expand=True)
# Add the columns to the new dataframe 'dfa' from the previous dataframe "df"
dfa["Impressions"] = df["Impressions"]
dfa["Clicks"] = df["Clicks"]
dfa["Complete_Views_Video"] = df["Complete Views (Video)"]
dfa["Audience List Type"] = df['Audience List']
```

```python
# Rename the column names
dfa.columns = ['Audience source', 'Categories', 'T1', 'T2', 4, 'Impressions', 'Clicks', 'Complete_Views_Video', 'Audience List Type']
# Remove the column no 4 which is not required
dfa.drop([4], axis = 1, inplace = True)
# Filter the Audience source for 'Display & Video 360 Audiences' and the Categories for 'Affinity Categories' and 'In-Market Categories'

dff = dfa[dfa['Audience List Type'] == 'Bid Manager Audiences']

# Change the datatype of metrics to integer
dff['Impressions'] = dff['Impressions'].astype(int)
dff['Clicks'] = dff['Clicks'].astype(int)
dff['Complete_Views_Video'] = dff['Complete_Views_Video'].astype(int)
# Filter out the rows with Impressions less than 1000
greateror1000 =  dff['Impressions'] > 999
dff = dff[greateror1000]

# Remove the null T2 value
dff.dropna(inplace = True)
# Create CTR and VCR Columns

dff['CTR'] = dff["Clicks"]/dff['Impressions']
dff['VCR'] = dff["Complete_Views_Video"]/dff['Impressions']
```

```python
# Save the output in a csv file
dff.to_csv(r"C:\\Users\\adishree.b\\Documents\\Files\\post campaign analysis and audience\\out3.csv", index = False)
```
